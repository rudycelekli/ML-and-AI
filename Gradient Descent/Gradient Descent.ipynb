```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
# Importing the data an
data = pd.read_csv('data.csv')
```

```python
#Taking a look at the data
data.head()
```

```python
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
plt.scatter(X, Y, alpha=0.75)
plt.show()
```

```python
def Error(m,b,data):
    totalError = 0
    x = data['x']
    y = data['y']
    for i in range(0,len(data)):
        error = (y[i] - (m*x[i] + b))**2
        totalError += error
    return totalError/ float(len(data))
```

```python
Error(0,0,data)
```

```python
# Defining the Gradient Descent function/algorithm
def gradient_descent(L,num_iterations):
    n = float(len(X)) # Number of elements in X
    m = random.randint(0,10) #random as starting with zero was not necessarily returning minimized error
    b = random.randint(0,10)
    m_start = m #so i can track performance depending on different initial values
    b_start = b
    error = 0
    errorList=[] #creating a list so i can record and plot the epochs vs error graph
    epochList=[] #creating a list so i can record and plot the epochs vs error graph
    for i in range(num_iterations): 
        Y_pred = m*X + b  # The current predicted value of Y
        D_m = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt m
        D_b = (-2/n) * sum(Y - Y_pred)  # Derivative wrt c
        m = m - L * D_m  # Update m
        b = b - L * D_b  # Update c
        error = Error(m,b,data)
        errorList.append(error)
        epochList.append(i)
        print ('epoch',i,'-- m:',m,'-- b:',b,'--Error:',error,'--initialm',m_start,'--initialb',b_start)
    plt.scatter(epochList, errorList) # epochs vs error
    plt.show()
    plt.scatter(X, Y, alpha=0.75) 
    plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
    plt.show()
```
